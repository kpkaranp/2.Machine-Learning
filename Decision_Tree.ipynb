```python
#importing the Libraies pandas, matplotlib.pylot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset      # verify data
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)   # dummies
```

```python
datasets
```

```python
indep=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep=datasets[['Profit']]    #spilt inpu and output
```

```python
#split into training set and test above data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 0.3, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor     # Model created with Decion tree algoritham
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best')
regressor=regressor.fit(X_train,y_train)      
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score #Evalutaion with R2
r_score=r2_score(y_test,y_pred)
```
